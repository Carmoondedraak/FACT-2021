```python
import sys
import os
import matplotlib.pyplot as plt
```

```python
# Read and split some result files
anomaly_results_dir = './results/anomaly-maps/'
img_names = os.listdir(anomaly_results_dir)
mvtec_objects = ['leather.png', 'hazelnut.png', 'wood.png', 'tile.png', 'metalnut.png', 'pill.png']
mvtec_files = [img_file for img_file in img_names if img_file in mvtec_objects]
mvtec_files = [os.path.join(anomaly_results_dir, file) for file in mvtec_files]
mnist_files = [img_file for img_file in img_names if 'model' in img_file]
mnist_files = [os.path.join(anomaly_results_dir, file) for file in mnist_files]
ucsd_files = [img_file for img_file in img_names if 'ucsd' in img_file]
ucsd_files = [os.path.join(anomaly_results_dir, file) for file in ucsd_files]
```

```python
# Function for showing results
def show_results(files, option='mvtec'):
    images = [plt.imread(file) for file in files]
    
    if option == 'mvtec':
        fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)
    elif option == 'mnist':
        fig, axes = plt.subplots(1, 2, sharex=True, sharey=True)
    elif option == 'ucsd':
        img = images[0]
        plt.imshow(img)
        return

    for img, ax in zip(images, axes.flat):
        ax.imshow(img)
        ax.axis('off')
```

#### To replicate the results (qualitative and quantitative) such as AUROC and IoU for the anomaly maps part of the paper, please run the tests.py script in expVAE_pl/ folder.

```python
# Here we show the paper qualitative results that we achieved on MNIST
show_results(mnist_files, 'mnist')
```

```python
# Here we show the paper qualitative results that we achieved on UCSD Pedestrians dataset
show_results(ucsd_files, 'ucsd')
```

```python
# Here we show the paper qualitative results that we achieved on MVTec-AD dataset
show_results(mvtec_files, 'mvtec')
```
